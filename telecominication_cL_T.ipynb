```python
import pandas as pd
```

```python
df=pd.read_csv("cleaned.csv")
```

```python
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
df['Churn'].info()
```

```python
X=df.drop('Churn',axis=1)
y=df['Churn']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from  sklearn.linear_model import LogisticRegression
from sklearn.tree  import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,f1_score,roc_auc_score
```

```python
lr=LogisticRegression()
dt=DecisionTreeClassifier()
```

```python
y_train.head()
```

```python
lr.fit(x_train,y_train)
dt.fit(x_train,y_train)
```

```python
log_p=lr.predict(x_test)
dt_p=dt.predict(x_test)
```

```python
log_a=accuracy_score(y_test,log_p)
dt_a=accuracy_score(y_test,dt_p)
log_f1=f1_score(y_test,log_p)
dt_f1=f1_score(y_test,dt_p)
```

```python
import numpy as np
np.unique(y_train)
```

```python
x_train.head()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Input
model=Sequential([Input(shape=(X.shape[1],)),Dense(units=128,activation='relu'),Dense(units=64,activation='relu'),Dense(units=1,activation='sigmoid')])
from tensorflow.keras.callbacks import EarlyStopping
early_stopping=EarlyStopping(patience=5,monitor='val_loss',restore_best_weights=True)
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,validation_split=0.2,epochs=100,callbacks=[early_stopping])
```

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print("Test accuracy:", test_acc)
print("Test loss:", test_loss)
```

```python
print(f"Accurcy for logistic regression:{log_a}")
print(f"Accurcy for decision tree:{dt_a}")
print(f"Accurcy for neural network:{test_acc}")
print(f"F1 score for logistic regression:{log_f1}")
print(f"F1 score for decision tree:{dt_f1}")
```
