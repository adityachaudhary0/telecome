```python
import pandas as pd
```

```python
df=pd.read_csv("cleaned.csv")
```

```python
df.head()
```

```python
df['Churn'].info()
```

```python
X=df.drop('Churn',axis=1)
y=df['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from  sklearn.linear_model import LogisticRegression
from sklearn.tree  import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,f1_score,roc_auc_score
```

```python
lr=LogisticRegression()
dt=DecisionTreeClassifier()
```

```python
y_train.head()
```

```python
lr.fit(x_train,y_train)
dt.fit(x_train,y_train)
```

```python
log_p=lr.predict(x_test)
dt_p=dt.predict(x_test)
```

```python
log_a=accuracy_score(y_test,log_p)
dt_a=accuracy_score(y_test,dt_p)
log_f1=f1_score(y_test,log_p)
dt_f1=f1_score(y_test,dt_p)
```

```python
print(f"Accurcy for logistic regression:{log_a}")
print(f"Accurcy for decision tree:{dt_a}")
print(f"F1 score for logistic regression:{log_f1}")
print(f"F1 score for decision tree:{dt_f1}")
```

```python
model_ev=pd.DataFrame({"Model":["Logistic Regression", "Decision Tree"],
                       "Acuracy":[log_a,dt_a],
                       "F1_score":[log_f1,dt_f1]})
model_ev
```
